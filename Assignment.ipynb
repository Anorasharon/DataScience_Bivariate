4)What kind of relation between salary and mba_p

```python

def quanqual(dataset):
    quan=[]
    qual=[]
    for j in dataset.columns:
        if (dataset[j].dtype=="O"):
            qual.append(j)
        else:
            quan.append(j)
    return quan,qual
```
